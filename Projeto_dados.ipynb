```python
#Importando bibliotecas
import pandas as pd
import plotly.express as px
```

```python
# Incluindo dataframes
df20 = pd.read_csv("/content/acidentes_2020-novo.csv", on_bad_lines="warn", sep = ";")
df21 = pd.read_csv("/content/acidentes2021.csv", on_bad_lines="warn", sep = ";")
df22 = pd.read_csv("/content/acidentes2022.csv", on_bad_lines="warn", sep = ";")
df23 = pd.read_csv("/content/acidentes2023.csv", on_bad_lines="warn", sep = ";")
df24 = pd.read_csv("/content/acidentes2024.csv", on_bad_lines="warn", sep = ";")
```

```python
#Iniciar análise com 2024
pd.set_option('display.max_columns', None)
df24.head()
```

```python
#Checar tipo das variáveis
df24.info()
```

```python
#Mudar tipo da data
df24['data']=pd.to_datetime(df24['data'])

# Substituir ponto por nada (milhares) e vírgula por ponto (decimal), depois converter para float
df24['vitimas'] = df24['vitimas'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['vitimasfatais'] = df24['vitimasfatais'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['auto'] = df24['auto'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['moto'] = df24['moto'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['ciclom'] = df24['ciclom'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['ciclista'] = df24['ciclista'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['pedestre'] = df24['pedestre'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['onibus'] = df24['onibus'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['caminhao'] = df24['caminhao'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['viatura'] = df24['viatura'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
df24['outros'] = df24['outros'].apply(lambda x: float(x.replace('.', '').replace(',', '.')))
pd.set_option('display.max_columns', None)
df24.head(2)
```
